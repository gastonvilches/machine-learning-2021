```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

Cargo los datos:

```python
x = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
df1 = pd.DataFrame({'x':x, 'y':[8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]})
df2 = pd.DataFrame({'x':x, 'y':[9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]})
df3 = pd.DataFrame({'x':x, 'y':[7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]})
df4 = pd.DataFrame({'x':x, 'y':[6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]})
```

Visualizo los datos:

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(15,10))
fig.suptitle('Cuarteto de Anscombe')

sns.scatterplot(ax=axes[0,0], x='x', y='y', data=df1)
axes[0,0].set_title('Dataset N°1')

sns.scatterplot(ax=axes[0,1], x='x', y='y', data=df2)
axes[0,1].set_title('Dataset N°2')

sns.scatterplot(ax=axes[1,0], x='x', y='y', data=df3)
axes[1,0].set_title('Dataset N°3')

sns.scatterplot(ax=axes[1,1], x='x', y='y', data=df4)
axes[1,1].set_title('Dataset N°4')
```

Agrego los 4 puntos a los dataframes con una etiqueta que los distinga del resto:

```python
df1['type'] = ['data'] * df1.shape[0]
df2['type'] = ['data'] * df2.shape[0]
df3['type'] = ['data'] * df3.shape[0]
df4['type'] = ['data'] * df4.shape[0]

points = pd.DataFrame({'x':[16, 8, 16, 16], 'y':[8, 6, 10, 6], 'type':['P1', 'P2', 'P3', 'P4']})

df1 = df1.append(points);
df2 = df2.append(points);
df3 = df3.append(points);
df4 = df4.append(points);
```

Ahora muestro los gráficos de nuevo añadiendo los 4 puntos con distinto color:

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True, figsize=(15,10))
fig.suptitle('Cuarteto de Anscombe')

sns.scatterplot(ax=axes[0,0], x='x', y='y', data=df1, hue='type')
axes[0,0].set_title('Dataset N°1')

sns.scatterplot(ax=axes[0,1], x='x', y='y', data=df2, hue='type')
axes[0,1].set_title('Dataset N°2')

sns.scatterplot(ax=axes[1,0], x='x', y='y', data=df3, hue='type')
axes[1,0].set_title('Dataset N°3')

sns.scatterplot(ax=axes[1,1], x='x', y='y', data=df4, hue='type')
axes[1,1].set_title('Dataset N°4')
```

Por inspección visual parecería que P4 pertenece al dataset N°2 y P3 pertenece al N°3. En los otros dos no es tan claro, pero parecería que P1 pertenece al N°4 y P2 al N°1. Si bien podría utilizar algún estadístico para ver cómo varía al agregar cada nuevo dato, no me parece muy apropiado porque en este caso los datasets tienen formas bien definidas y los estadísticos apropiados para cada forma serían diferentes. Por ejemplo, el r^2 sería apropiado para el dataset N°3 pero no para el N°2.

La conclusión aquí sería que los datasets pueden tener formas variadas que a veces los estadísticos descriptivos que usamos no las representan, y que siempre es útil visualizar los datos porque se pueden distinguir patrones y formas que sería difícil de describir matemáticamente.
